```python
from simple_linear_regression import SimpleLinearRegression, scaler
import matplotlib.pyplot as plt
```

```python
train_x = [1,4,2,5,7,4,6]
train_y = [0.5,2,1,2.5,3.5,2,3]
test_x = [3,6,2,10,46,23,63,137,111]
```

## Without Using Scaler

```python
model = SimpleLinearRegression(num_epochs=10).fit(train_x, train_y)
```

```python
model.predict(test_x)
```

```python
model.loss_history
```

```python
plt.plot(model.loss_history,marker="x", color="red")
plt.xlabel("No of Epochs")
plt.ylabel("Error")
```

```python
model1 = SimpleLinearRegression(num_epochs=1000).fit(train_x,train_y)
```

```python
model1.predict(test_x)
```

```python
model1.loss_history
```

```python
plt.plot(model1.loss_history,marker="x", color="red")
plt.xlabel("No of Epochs")
plt.ylabel("Error")
```

## SGD (Batch Size 1)

```python
model2 = SimpleLinearRegression(num_epochs=500).fit(train_x, train_y, batch_size=1)
```

```python
model2.predict(test_x)
```

```python
model2.loss_history
```

```python
plt.plot(model2.loss_history,marker="x", color="red")
plt.xlabel("No of SGD Batches")
plt.ylabel("Error")
```

## Using Scaler

```python
x_train_scaled, min_x, max_x = scaler(train_x)
```

```python
scaled_trained_model = SimpleLinearRegression(num_epochs=10).fit(x_train_scaled, train_y)
```

```python
x_test_scaled, _, _ = scaler(test_x, min_x, max_x)
x_test_scaled
```

```python
scaled_trained_model.predict(x_test_scaled)
```

```python
scaled_trained_model.loss_history
```

```python
plt.plot(scaled_trained_model.loss_history,marker="x", color="red")
plt.xlabel("No of Epochs")
plt.ylabel("Error")
```

```python
scaled_trained_model1 = SimpleLinearRegression(num_epochs=1000).fit(x_train_scaled, train_y)
```

```python
scaled_trained_model1.predict(x_test_scaled)
```

```python
scaled_trained_model1.loss_history
```
